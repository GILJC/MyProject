```python
# 과적합 이해하기
# 과적합(overfitting)이란 모델이 학습 데이터셋 안에서는 일정 수준 이상의 예측 정확도를 보이지만, 
# 새로운 데이터에 적용하면 잘 맞지 않는 것을 말합니다.

# 학습은 rainset으로 시키돼,
# 마지막 evaluate는, testset으로 해야 한다는 것
```

```python

# 실제로 testset과 trainset을 구분해서 해보자, 라는 것
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
# seed 값 설정
seed = 0
np.random.seed(seed)
tf.random.set_seed(3)
```

```python
df = pd.read_csv('data/sonar.csv')
```

```python
print(df.head())
```

```python
dataset = df.values
X = dataset[:, :60].astype(float)
Y_obj = dataset[:, 60]
```

```python
e = LabelEncoder()
e.fit(Y_obj)
Y = e.transform(Y_obj)  # 결과 Y 는 transform 해줌
```

```python
# from sklearn.model_selection import train_test_split
# 학습셋과 테스트셋의 구분
X_train, X_test, y_train, y_test = \
    train_test_split(X, Y, test_size = 0.3, random_state = seed)  # X,Y주고, test data는 30% 로 쪼갬
```

```python
# 아까와 똑같은 모델을 그대로 씀
model = Sequential()
model.add(Dense(24, input_dim = 60, activation = 'relu'))
model.add(Dense(10, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
# 이것도 똑같이
model.compile(loss = 'mean_squared_error', optimizer = 'adam', metrics = ['accuracy'])
```

```python
# fit은 epochs를 줄임  200 -> 130
model.fit(X_train, y_train, epochs = 130, batch_size = 5)  # X,Y도 직접 주는게 아니라
# train을 주는 것
```

```python
# 테스트셋에 모델 적용
print("Test Accuracy : %.4f" % (model.evaluate(X_test, y_test)[1]))
# evaluate에 test data를 넣음
```
