## 활성화 함수(Activation function)

#### 01. Sigmoid
- y = 1 / (1 + e<sup>-x</sup>)

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid_function(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-5, 5)  # x 값을 np의 linspace를 이용해서 (-5에서 5까지)
y = sigmoid_function(x)
plt.plot(x, y)
plt.title("Sigmoid")
plt.show()

# sigmoid 함수 = 시그모이드 함수는 S자형 곡선 또는 시그모이드 곡선을 갖는 수학 함수
# 시그모이드 함수는 실수 전체를 정의역으로 가지며, 반환값은 단조증가하는 것이 일반적이지만
# 단조감소할 수도 있다.
# 시그모이드 함수의 반환값(y축)은 흔히 0에서 1까지의 범위를 가진다.
```

```python
# sigmoid함수의 도함수
```

#### 02. tanh
- y = (e<sup>x</sup> - e<sup>-x</sup>) / (e<sup>x</sup> + e<sup>-x</sup>)

```python
# 이건 하이퍼볼릭 탄젠트라고 불림
# 딥 러닝 라이브러리가 내장하고있기떄문에 구현하진 않고, 
# 메커니즘을 알기 위해 함수를 한번 구현해 볼 것
```

```python
def tanh_function(x):
    return np.tanh(x)

x = np.linspace(-5, 5)
y = tanh_function(x)
plt.plot(x, y)
plt.title("tanh")
plt.show()
```

```python
# sigmoid와 같지만, 
# 얘는 -1 과 1사이를 왔다갔다 함
```

#### 03. ReLU(Rectified Linear Unit)
- y = 0 if (x <= 0) else x

```python
# x가 0보다 작거나 같은때는 0, 아닐 때는 x  (=y값)
```

```python
def relu_function(x):
    return np.where(x <= 0, 0, x)  # 얘가 나중에 스칼라값이 아니라 matrix로도 들어올 수 있으므로, 넘파이(np)의 도움을 받아야 함

x = np.linspace(-5, 5)
y = relu_function(x)
plt.plot(x, y)
plt.title("relu")
plt.show()
```

```python
# 이게 relu
```

#### 04. Leaky ReLU
- X가 음수인 영역에서 아주 작은 기울기를 생성
- 출력이 0이 되어 더 이상 학습이 진행도지 않는 뉴런이 발생하는 dying ReLU 현상을 피하기 위해 사용 ( = 음수인 영역에서 약간의 기울기를 준다는 소리(?) )
- y = 0.01*x if (x <= 0) else x

```python
def leaky_relu_function(x):
    return np.where(x <= 0, 0.01*x, x)

x = np.linspace(-5, 5)
y = leaky_relu_function(x)
plt.plot(x, y)
plt.title("Leaky relu")
plt.show()
```

```python
# 미세하지만 음수쪽에 기울기가 살아 있는 것
```

#### 05. Softmax
- 입력받은 값을 출력으로 0 ~ 1 사이의 값으로 모두 정규화하며, 출력 값들의 총합은 항상 1이 되는 특성을 가진 함수


```python
def softmax_function(x):
    return np.exp(x) / np.sum(np.exp(x))

y = softmax_function(np.array([1, 2, 3]))
print(y)
```

```python
# 다 더하면 항상 1이 되게 되어있음
```
