```python
def preprocessing(review, remove_stopwords=False):
    review_text = BeautifulSoup(review, 'lxml').get_text()
    review_text = re.sub(r"[^A-Za-z]", ' ', review_text)
    words = review_text.lower().split()
    
    if remove_stopwords:
        stops = set(stopwords.words('english'))
        words = [w for w in words if not w in stops]
        clean_review =  ' '.join(words)  # 다시 붙임
    else:  # 만약 불용어 제거하라는 옵션이 안들어오면,
        clean_review =  ' '.join(words)  # 다시 붙임,   이것만 함
        
    return clean_review
```

```python
clean_train_reviews = []
for review in train_data['review']:
    clean_train_reviews.append(preprocessing(review, remove_stopwords=True))
```

```python
# 제대로 됐는지 하나만 찍어봄
print(clean_train_reviews[0])
```

```python
# 일단은 이렇게 필요한 전처리를 끝냈음

# 이제 index를 벡터로 만들어야함
# 모델을 만들려면 길이가 동일해야함 -> 패딩 작업 할 것
```

```python
# 일단 지금까지 한걸 pandas 데이터로 만들어 두자(전처리한 데이터)
clean_train_df = pd.DataFrame({'review': clean_train_reviews,
                              'sentiment': train_data['sentiment']}) #원래 데이터에 sentiment 해서 dataframe 만듬
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(clean_train_reviews)
text_sequences = tokenizer.texts_to_sequences(clean_train_reviews)
```

```python
print(text_sequences[0])  # 첫번째 review가 벡터화된걸 확인

# 위와 같이, 각 리뷰가 text가 아닌 index의 벡터로 구성됨
```

```python
word_vocab = tokenizer.word_index
```

```python
print(type(word_vocab))
```

```python
# word_vocab   # 단어 사전 만든 것 확인
# 너무 기니까 지워두자
```

```python
print('전체 단어 수:', len(word_vocab))
```

```python
data_config = {}
data_config['vocab'] = word_vocab
data_config['vocab_size'] = len(word_vocab)
```

```python
# 리뷰 단어 수 중간값: 174.0
# 중간값이 174.  이걸 기준으로 할거임  ->  oulier의 영향을 최대한 피하려고
# => 평균값은 영향을 받아서 중간값을 사용

MAX_SEQUENCE_LENTH = 174
train_inputs = pad_sequences(text_sequences, maxlen=MAX_SEQUENCE_LENTH,
                           padding='post')
# 이놈보다 길이가 긴놈은 짜르고, 패딩=포스트로 뒤쪽에다 붙임

print('Shape of train data:', train_inputs.shape)
```

```python
train_labels = np.array(train_data['sentiment'])
print('Shape of label tensor:', train_labels.shape)
```

- 총 4개의 데이터를 저장\
a. 정제된 텍스트 데이터\
b. 벡터화한 데이터\
c. 정답 레이블\
d. 데이터 정보(단어 사전, 전체 단어 수)

```python
DATA_IN_PATH = './data_in/'   # 폴더 안에 저장하기로 위치 지정
TRAIN_INPUT_DATA = 'train_input.npy'  # 공식적으로 이런 확장자는 없지만, 햇갈리지 않으려고 일부러 이렇게 씀
TRAIN_LABEL_DATA = 'train_label.npy'
TRAIN_CLEAN_DATA = 'train_clean.csv'   # 전처리 끝난 데이터
DATA_CONFIGS = 'data_configs.json'  # 아까 dict 만든 것 -> dict는 json 형식으로 바로 변환이 됨

import os
if not os.path.exists(DATA_IN_PATH):   # DATA_IN_PATH  <-  이게 존재하지 않으면
    #만들어라
    os.makedirs(DATA_IN_PATH)
    
# np로 만든건 np가 저장기능이 있음.
np.save(open(DATA_IN_PATH + TRAIN_INPUT_DATA, 'wb'), train_inputs)  # wb -> 바이너리로 저장하라
np.save(open(DATA_IN_PATH + TRAIN_LABEL_DATA, 'wb'), train_labels)

# 처음에 벡터화 하기 전 text 데이터 -> dataframe으로 만들어 놓음
clean_train_df.to_csv(DATA_IN_PATH + TRAIN_CLEAN_DATA, index=False)   # '+ ~~' -> 파일명

json.dump(data_config, open(DATA_IN_PATH + DATA_CONFIGS, 'w'), ensure_ascii=False)  #  open(위치 + 파일명)
```

```python
test_data = pd.read_csv(DATA_IN_PATH + "testData.tsv", header=0, \
                        delimiter="\t", quoting=3)

clean_test_reviews = []
for review in test_data['review']:
    clean_test_reviews.append(preprocessing(review, remove_stopwords=True))


clean_test_df = pd.DataFrame({'review': clean_test_reviews, 'id': test_data['id']})
test_id = np.array(test_data['id'])

text_sequences = tokenizer.texts_to_sequences(clean_test_reviews)
test_inputs = pad_sequences(text_sequences, maxlen=MAX_SEQUENCE_LENTH, \
                            padding='post')             # 원래 LENGTH가 맞는데 #오타라서 그냥 LENTH로 씀
```

```python
# 이제 저장하면 됨
# 코드
TEST_INPUT_DATA = 'test_input.npy'
TEST_CLEAN_DATA = 'test_clean.csv'
TEST_ID_DATA = 'test_id.npy'

np.save(open(DATA_IN_PATH + TEST_INPUT_DATA, 'wb'), test_inputs)
np.save(open(DATA_IN_PATH + TEST_ID_DATA, 'wb'), test_id)
clean_test_df.to_csv(DATA_IN_PATH + TEST_CLEAN_DATA, index=False)
```
