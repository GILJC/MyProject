```python
import os
import json
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras import layers
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
def plot_graphs(history, string):
    plt.plot(history.history[string])
    plt.plot(history.history['val_' + string], '')
    plt.xlabel('Epochs')
    plt.ylabel(string)
    plt.legend([string, 'val_' + string])
    plt.show()
```

```python
DATA_IN_PATH = './data_in/'
DATA_OUT_PATH = './data_out/'
TRAIN_Q1_DATA_FILE = 'train_q1.npy'
TRAIN_Q2_DATA_FILE = 'train_q2.npy'
TRAIN_LABEL_DATA_FILE = 'train_label.npy'
DATA_CONFIGS = 'data_configs.json'
```

```python
SEED_NUM = 1234
tf.random.set_seed(SEED_NUM)
```

```python
# data 로드(load)
q1_data = np.load(open(DATA_IN_PATH + TRAIN_Q1_DATA_FILE, 'rb'))
q2_data = np.load(open(DATA_IN_PATH + TRAIN_Q2_DATA_FILE, 'rb'))
labels = np.load(open(DATA_IN_PATH + TRAIN_LABEL_DATA_FILE, 'rb'))
prepro_configs = json.load(open(DATA_IN_PATH + DATA_CONFIGS, 'r'))
```

```python
model_name = 'cnn_similarity'
BATCH_SIZE = 1024
NUM_EPOCHS = 100
VALID_SPLIT = 0.1
MAX_LEN = 31

kargs = {'model_name': model_name,
        'vocab_size': prepro_configs['vocab_size'],
        'word_embedding_dimension': 100,
        'conv_num_filters': 300,
        'conv_windows_size': 3,
        'max_pool_seq_len': MAX_LEN,
        'sent_embedding_dimension': 128,
        'dropout_rate': 0.2,
        'hidden_dimension': 200,
        'output_dimension': 1}
```

```python
# 일부 중간 레이어만 만들것
class SentenceEmbedding(layers.Layer):  # 레이어로 받음
    def __init__(self, **kargs):
        super(SentenceEmbedding, self).__init__()
        
        # 그리고 안에 layer를 정의함
        self.conv = layers.Conv1D(kargs['conv_num_filters'],  # ['출력갯수'] 
                                  kargs['conv_windows_size'],  # ['커널사이즈'] 
                                  activation='relu', padding='same')
        self.max_pool = layers.MaxPool1D(kargs['max_pool_seq_len'], 1)  # 풀사이즈는 얘로 잡음
        # 전채에서 하나 뽑겠다는 것
        
        # 그 다음 dense 층
        self.dense = layers.Dense(kargs['sent_embedding_dimension'], activation='relu')   # dense 층 출력은 이거.
        
        # 그리고  call 함수
        def call(self, x):
            x = self.conv(x)
            x = self.max_pool(x)
            x = self.dense(x)
            
            return tf.squeeze(x, 1)
```

```python
class SentenceSimilarityModel(tf.keras.Model):
    def __init__(self, **kargs):
        super(SentenceSimilarityModel, self).__init__(name=kargs['model_name'])
        
        self.word_embedding = layers.Embedding(kargs['vocab_size'] + 1,  # 임베딩 출력은 vocab_size, 입력은 +1
                                               kargs['word_embedding_dimension'])  # 출력은 word_embedding_dimension
        self.base_encoder = SentenceEmbedding(**kargs)
        self.hypo_encoder = SentenceEmbedding(**kargs)
        # 2개를 만들거임
        
        self.dense = layers.Dense(kargs['hidden_dimension'], activation='relu')  # 첫번째 dense층의 출력은 hidden_dimension
        # self.logit = layers.Dense(1, activation='sigmoid')   # 출력은 1  -> 에서 변경 -> 만들어놨으니 -> 일관성있께 kargs로 쓰자
        self.logit = layers.Dense(kargs['output_dimension'], activation='sigmoid')
        self.dropout = layers.Dropout(kargs['dropout_rate'])
        
    def call(self, x):
        # 입력이 들어오면, x1, x2로 언패킹 하는 것
        x1, x2 = x
        b_x = self.word_embedding(x1)   # 각각을 이렇게 통과시킴
        h_x = self.word_embedding(x2)
        # 그렇게 나온 결과를 b_x, h_x에 넣음
        
        # 이상태에서 각각을 dropout 시킬 것
        b_x = self.dropout(b_x)
        h_x = self.dropout(h_x)
        # 질문 2개를 각각 처리하고 있음
        
        b_x = self.base_encoder(b_x)
        h_x = self.hypo_encoder(h_x)
        # 책의 그림처럼 쪼개서 각각 집어 넣은 것 (p260)
        
        # 1차원으로 
        e_x = tf.concat([b_x, h_x], -1)
        e_x = self.dense(e_x)
        e_x = self.dropout(e_x)
        
        # 그렇게 나온 결과를 맨 마지막 층에 통과시키면 됨
        return self.logit(e_x)
    
    # 이러면 끝.  그림과 한번 비교해 볼 것
```

```python
# 모델 생성
model = SentenceSimilarityModel(**kargs)
model.compile(optimizer=tf.keras.optimizers.Adam(1e-3),
             loss=tf.keras.losses.BinaryCrossentropy(),
             metrics=[tf.keras.metrics.BinaryAccuracy(name='accuracy')])
```

```python
earlystop_callback = EarlyStopping(monitor='val_accuracy', min_delta=0.0001, patience=3)\

checkpoint_path = DATA_OUT_PATH + model_name + '/weights.h5'
checkpoint_dir = os.path.dirname(checkpoint_path)

if os.path.exists(checkpoint_dir):
    print('{} -- Folder already exists \n'.format(checkpoint_dir))
else:
    os.makedirs(checkpoint_dir, exist_ok=True)  # 얘를 빼서 dir이 안만들어졌었음, 이거 코딩 다시해서 돌림
    print('{} -- Folder create complete \n'.format(checkpoint_dir))
    
cp_callback = ModelCheckpoint(checkpoint_path, monitor='val_accuracy', verbose=1,
                             save_best_only=True, save_weights_only=True)
```

```python
history = model.fit((q1_data, q2_data), labels, batch_size=BATCH_SIZE,  # q1_data, q2_data 를 튜플로 줌
                    epochs=NUM_EPOCHS, validation_split=VALID_SPLIT,
                   callbacks=[earlystop_callback, cp_callback])
```

```python
plot_graphs(history, 'accuracy')
```

```python
plot_graphs(history, 'loss')
```
