```python
city = ['서울', '인천', '대전', '대구', '울산', '부산', '광주']
lat  = [37.56, 37.45, 36.35, 35.87, 35.53, 35.18, 35.16]
lon = [126.97, 126.70, 127.38, 128.60, 129.31, 129.07, 126.85]  # 위 5개 도시의 위도, 경도
pop_den = [16154, 2751, 2839, 2790, 1099, 4454, 2995]  # 인구 밀도
# 얘를 시각화를 해봅시다
```

```python
member_IDs = ['m_01', 'm_02', 'm_03', 'm_04']
before_ex = [27, 35, 40, 33]
after_ex = [30, 38, 42, 37]

n_data = len(member_IDs)
index = np.arange(n_data)

#   --- 이부분 안쓰면 한글 다 꺠져서 나옴 ---
# 
# import platform
# from matplotlib import font_manager, rc # Runtime Configuration
# 
# path = "c:/Windows/Fonts/malgun.ttf"  # Windows 밑에있는 맑은 폰트
# if platform.system() == 'Darwin':  
#     rc('font', family='AppleGothic')   # 맥일 경우 애플 고딕
# elif platform.system() == 'Windows':  # Window면 
#     font_name = font_manager.FontProperties(fname=path).get_name()   #  path를 쓰겠다는 것
#     rc('font', family=font_name)
# else:
#     print('Unknown system... sorry~~~~')   # 이도저도 아니면 에러
# 
# 

barWidth = 0.4
plt.bar(index, before_ex, color='c', align='edge',  # align=edge를 주면 (다음 그래프가)바로 오른쪽에 딱 붙어서 그려짐 -> 안주면 다음 그래프에 붙어서 모양이 이상해짐
        width=barWidth, label='before')
plt.bar(index + barWidth, after_ex, color='m', align='edge',
       width=barWidth, label='after')
plt.xticks(index + barWidth, member_IDs)  #  barWidth  -> 안쓰면 m_01, m_02, m_03...들이 가운데에 안오고 왼쪽에 붙음(+barWidth만큼 오른쪽으로 옮겨줌)
plt.legend()
plt.xlabel('회원ID')
plt.ylabel('윗몸 일으키기')
plt.title('운동 시작 전/후 근지구력 변화 비교')
plt.show()
```

```python
fruit = ['사과', '바나나', '딸기', '오렌지', '포도']
result = [7, 6, 3, 2, 2] # 사과 7명, 바나나 6명, 딸기 3명, 오렌지, 포도 2명
# 이걸 pie 그래프로 그릴 것
plt.pie(result)
plt.show()

# default 상태의 pie 그래프는 파란색 오른쪽 끝(3시방향) 이 시작점이고, 반시계방향(왼쪽방향)으로 돌면서 채움 (파랑->주황->녹색->빨강->보라)
```

```python
plt.pie(result, labels=fruit, autopct='%.1f%%')  # 소수점 아래 1번쨰까지 나타내고, 뒤에 %기호를 붙인다 (% 2개 붙이면 실제 %를 붙임)
plt.show()

# fruit = ['사과', '바나나', '딸기', '오렌지', '포도']  -> lebel
# result = [7, 6, 3, 2, 2]    ->  퍼센테이지(그레프의 차지하는 양)
```

```python
# default는 3시방향에서 시작 => 좌표상에서 0도가 되는 부분.
# 12시방향(좌표상 90도) 에서 시작되게 하고   /   시계방향으로 채워지게 해보자

plt.pie(result, labels=fruit, autopct='%.1f%%', startangle=90,  # startangle=90  -> (좌표상 90도=12시방향 에서 시작)
       counterclock=False)   #  counterclock=False  ->  시계 반대방향을 False로 =>  시계방향으로.
plt.show()
```

```python
# 강조
explode_value = [0.1, 0, 0, 0, 0]   #일단 tuple 데이터를 만듬   
plt.pie(result, labels=fruit, autopct='%.1f%%', startangle=90,\
       counterclock=False, explode=explode_value, shadow=True)  # explode를 조절해서 더 멀리 나가게 할 수도 있음.   /  shadow 로 그림자 생성
plt.show()

# 우리 유치원생들은 사과를 제일 좋아해요 -> 라는 뜻
```
