```python
import pandas as pd
```

```python
def processing_tool(file_name):
    file1_1 = pd.read_excel(file_name, sheet_name='(1)', header=10)  # 위에 10줄 지우고 가져오기

    file1_1.rename(columns={file1_1.columns[0]:'업종명'}, inplace=True)
    file1_1.rename(columns={file1_1.columns[1]:'1'}, inplace=True)
    file1_1.rename(columns={file1_1.columns[2]:'2'}, inplace=True)
    file1_1.rename(columns={file1_1.columns[3]:'총사업자_총계'}, inplace=True)
    file1_1.rename(columns={file1_1.columns[4]:'총사업자_신규'}, inplace=True)
    file1_1.rename(columns={file1_1.columns[5]:'총사업자_폐업'}, inplace=True)
    file1_1.rename(columns={file1_1.columns[6]:'6'}, inplace=True)
    file1_1.rename(columns={file1_1.columns[7]:'7'}, inplace=True)
    file1_1.rename(columns={file1_1.columns[8]:'8'}, inplace=True)

    file1_1.drop(['1', '2', '6', '7', '8'], axis=1, inplace=True)

    file1_1 = file1_1.drop(index=[24, 25, 26, 27, 28, 29], axis=0) # axis=1 -> 새로 지우기 / axis=0 -> 가로 값 지우기

    file1_1.set_index('업종명', inplace=True)

    file1_1=file1_1.transpose()

    # \n 삭제.

    file1_1.rename(columns={file1_1.columns[4]:'음식료품 및 담배제조업'}, inplace=True)
    file1_1.rename(columns={file1_1.columns[6]:'음식료품 및 담배제조업'}, inplace=True)
    file1_1.rename(columns={file1_1.columns[7]:'가죽, 가방 및 신발류 제조업'}, inplace=True)
    file1_1.rename(columns={file1_1.columns[8]:'목재 및 나무제품 제조업(가구제외)'}, inplace=True)
    file1_1.rename(columns={file1_1.columns[9]:'펄프, 종이 및 종이제품 제조업'}, inplace=True)
    file1_1.rename(columns={file1_1.columns[10]:'출판, 인쇄 및기록매체 제조업'}, inplace=True)
    file1_1.rename(columns={file1_1.columns[11]:'석탄, 석유정제품 및 핵연로 제조업'}, inplace=True)
    file1_1.rename(columns={file1_1.columns[12]:'화합물 및 화학 제품 제조업'}, inplace=True)
    file1_1.rename(columns={file1_1.columns[13]:'고무 및 프라스틱 제품 제조업'}, inplace=True)
    file1_1.rename(columns={file1_1.columns[16]:'조립금속제품 제조업(기계 및 장비 제외)'}, inplace=True)
    file1_1.rename(columns={file1_1.columns[17]:'기타기계 및 장비 제조업'}, inplace=True)
    file1_1.rename(columns={file1_1.columns[18]:'사무, 계산 및 회계용 기계 조립업'}, inplace=True)
    file1_1.rename(columns={file1_1.columns[19]:'기타 전지기계 및 전기변환장치 제조업'}, inplace=True)
    file1_1.rename(columns={file1_1.columns[20]:'영상, 음향 및 통신장비 제조업'}, inplace=True)
    file1_1.rename(columns={file1_1.columns[21]:'의료, 정밀, 광학기기 및 시계제조업'}, inplace=True)
    file1_1.rename(columns={file1_1.columns[22]:'자동차 및 트레일러 제조업'}, inplace=True)

    file1_1=file1_1.transpose()

    #-------------------------------------------------------------- part1  clear

    file1_2 = pd.read_excel(file_name, sheet_name='(3)', header=9)  # 위에 9줄 지우고 가져오기

    file1_2.rename(columns={file1_2.columns[0]:'업종명'}, inplace=True)
    file1_2.rename(columns={file1_2.columns[1]:'1'}, inplace=True)
    file1_2.rename(columns={file1_2.columns[2]:'2'}, inplace=True)
    file1_2.rename(columns={file1_2.columns[3]:'총사업자_총계'}, inplace=True)
    file1_2.rename(columns={file1_2.columns[4]:'총사업자_신규'}, inplace=True)
    file1_2.rename(columns={file1_2.columns[5]:'총사업자_폐업'}, inplace=True)
    file1_2.rename(columns={file1_2.columns[6]:'6'}, inplace=True)
    file1_2.rename(columns={file1_2.columns[7]:'7'}, inplace=True)
    file1_2.rename(columns={file1_2.columns[8]:'8'}, inplace=True)

    file1_2.drop(['1', '2', '6', '7', '8'], axis=1, inplace=True)

    file1_2 = file1_2.drop(index=[28, 29], axis=0) # axis=1 -> 새로 지우기 / axis=0 -> 가로 값 지우기

    file1_2.set_index('업종명', inplace=True)

    file1_2=file1_2.transpose()

    file1_2.rename(columns={file1_2.columns[11]:'여행알선 및 운수관련 서비스업'}, inplace=True)
    file1_2.rename(columns={file1_2.columns[13]:'금융 및 금융관련서비스업'}, inplace=True)
    file1_2.rename(columns={file1_2.columns[14]:'보험 및 보험관련서비스업'}, inplace=True)
    file1_2.rename(columns={file1_2.columns[16]:'기계장비 등 장비 임대업'}, inplace=True)
    file1_2.rename(columns={file1_2.columns[17]:'IT관련 및 연구개발업'}, inplace=True)
    file1_2.rename(columns={file1_2.columns[18]:'법무, 회계, 건축 및 상담업'}, inplace=True)
    file1_2.rename(columns={file1_2.columns[19]:'광고업 및 기타 산업관련 서비스업'}, inplace=True)
    file1_2.rename(columns={file1_2.columns[23]:'오락, 문화, 운동관련 산업 및 수리업'}, inplace=True)

    file1_2=file1_2.transpose()

    file1 = pd.concat([file1_1,file1_2])
    
    #----------------------------------------------------------------------part2 합침 완료.

    file1 = file1.reset_index()

    file1 = file1.reindex(columns=['업종명', '총사업자_총계','총사업자_신규','총사업자_폐업'])

    file_sum_all = file1['총사업자_총계'].sum()
    file_sum_new = file1['총사업자_신규'].sum()
    file_sum_close = file1['총사업자_폐업'].sum()

    file1_sum = {'업종명': '합계', '총사업자_총계': file_sum_all, '총사업자_신규': file_sum_new, '총사업자_폐업': file_sum_close}

    file1 = file1.append(file1_sum, ignore_index=True)  # ignore_index=Truen <- 없으면 에러남

    #file1

    file1.set_index('업종명', inplace=True)

    file1=file1.transpose()

    file1 = file1.reset_index()

    file1 = file1.reset_index()

    file1.rename(columns={file1.columns[0]:'0'}, inplace=True)

    file1.rename(columns={file1.columns[0]:'연도'}, inplace=True)
    file1.rename(columns={file1.columns[1]:'구분'}, inplace=True)

    file1["연도"] = a

    file1
    
    return file1
```
