```python
df2 = df[['연도', '사업자수']]
```

```python
df_total = df2.dropna()
df_total.reset_index(drop=True, inplace=True)
```

```python
df_opened = df_total.loc[[1, 4, 7, 10, 13, 16, 19, 22, 25, 28], :]
df_closed = df_total.loc[[2, 5, 8, 11, 14, 17, 20, 23, 26, 29], :]
df_total = df_total.loc[[0, 3, 6, 9, 12, 15, 18, 21, 24, 27], :]
```

```python
df_opened.reset_index(drop=True, inplace=True)
df_closed.reset_index(drop=True, inplace=True)
df_total.reset_index(drop=True, inplace=True)
```

```python
df_total.set_index('연도')
df_opened.set_index('연도')
df_closed.set_index('연도')
```

```python
df_all = pd.merge(df_total, df_opened, on='연도')
```

```python
df_all = pd.merge(df_all, df_closed, on='연도')
```

```python
df_all.columns = ['연도', '사업자수','신규','폐업']
```

```python
df_all.set_index('연도', inplace=True)
```

```python
df_all
```

```python
year = [i for i in range(2011, 2021)]
```

```python
# 총 사업자 수

plt.figure(figsize=(15,7))
plt.plot(df_all[['사업자수']], 'ro-')
plt.grid()
plt.title('연도별 사업자 수', fontsize=25)
plt.xticks(year, fontsize=13)
plt.yticks(fontsize=13)
plt.xlabel('연도', fontsize=17)
plt.ylabel('사업자 수 (백만명)', fontsize=17)
plt.legend(['총 사업자 수'], fontsize=15)
plt.show()
```

```python
# 연도별 신규 & 폐업 사업자 수

plt.figure(figsize=(15,7))
plt.plot(df_all[['신규','폐업']], 'o-')
plt.grid()
plt.title('연도별 신규 및 폐업 사업자 수', fontsize=25)
plt.xticks(year, fontsize=13)
plt.yticks(fontsize=13)
plt.xlabel('연도', fontsize=17)
plt.ylabel('사업자 수 (백만명)', fontsize=17)
plt.legend(['신규', '폐업'], fontsize=15)
plt.show()
```

```python
df3_cut
```

```python
df3_total = df3_cut.loc[[0, 3, 6], :]
df3_opened = df3_cut.loc[[1, 4, 7], :]
df3_closed = df3_cut.loc[[2, 5, 8], :]
```

```python
df3_sum = pd.merge(df3_opened, df3_closed, on='연도')
```

```python
df3_sum = df3_sum[2::-1]
```

```python
df3_sum.set_index('연도')
```

```python
df_list = []
for i in names:
    df_tmp = df3_sum[['연도', f'{i}_x', f'{i}_y']]
    df_tmp.set_index('연도', inplace=True)
    df_list.append(df_tmp.transpose())
```

```python
df_list
```

```python
names.index('음식점업')
```

```python
# 음식점업 신규 & 폐업률

x = np.array(range(len(df_list[33].columns)))
w = 0.3
plt.figure(figsize=(5,5))
plt.bar(x-0.15, df_list[33].loc['음식점업_x'], width=w, label='신규')
plt.bar(x+0.15, df_list[33].loc['음식점업_y'], width=w, label='폐업')
plt.xticks(range(3), df_list[33].columns, fontsize=13)
plt.yticks(fontsize=13)
plt.legend(loc=4, fontsize=10)
plt.xlabel('연도', fontsize=13)
plt.ylabel('사업자 수', fontsize=13)
plt.grid()
plt.show()
```

```python
# 51업종 신규 & 폐업률

plt.figure(figsize=(23, 80))

for i in range(len(df_list)):
    plt.subplot(13, 4, i+1)
    plt.title(f'{names[i]}\n신규 및 폐업 사업자 수 (명)', fontsize=15)
    x = np.array(range(len(df_list[i].columns)))
    plt.bar(x-0.15, df_list[i].loc[f'{names[i]}_x'], width=0.3, label=f'신규')
    plt.bar(x+0.15, df_list[i].loc[f'{names[i]}_y'], width=0.3, label=f'폐업')
    plt.xticks(range(3), df_list[i].columns, fontsize=13)
    plt.yticks(fontsize=13)
    plt.legend(loc=4, fontsize=10)
    plt.grid()
    
plt.show()
```

```python
pct1 = df3_total_pct.iloc[:, 0:10]
pct2 = df3_total_pct.iloc[:, 10:20]
pct3 = df3_total_pct.iloc[:, 20:30]
pct4 = df3_total_pct.iloc[:, 30:40]
pct5 = df3_total_pct.iloc[:, 40:]

pct = [df3_total_pct.iloc[:, i*10:i*10+10] for i in range(4)]
pct.append(df3_total_pct.iloc[:, 40:])
```

```python
# test

x = np.array(range(len(pct1.columns)))
w = 0.3
plt.figure(figsize=(20,8))
for i in range(len(pct)):    
plt.bar(x-0.15, pct1.loc[2019], width=w, label='2019')
plt.bar(x+0.15, pct1.loc[2020], width=w, label='2018')
plt.xticks(range(10), pct1.columns, rotation=15, fontsize=15)
plt.yticks(fontsize=15)
plt.title('업종별 전년 대비 증감률 (%)', fontsize=25)
plt.xlabel('업종 이름', fontsize=15)
plt.ylabel('전년 대비 증감률 (%)', fontsize=15)
plt.legend(fontsize=15)
plt.show()
```

```python
# 51업종 전년 대비 사업자 수 증감률

w = 0.3
plt.figure(figsize=(20,70))
for i in range(4):
    plt.subplot(5,1,i+1)
    x = np.array(range(len(pct[i].columns)))
    if i == 0:
        plt.title('업종별 전년 대비 사업자 수 증감률 (%)', fontsize=30)
    plt.bar(x-0.15, pct[i].loc[2019], width=w, label='2019')
    plt.bar(x+0.15, pct[i].loc[2020], width=w, label='2018')
    plt.xticks(range(10), pct[i].columns, rotation=25, fontsize=13)
    plt.yticks(fontsize=20)
    plt.ylabel('전년 대비 증감률 (%)', fontsize=20)
    plt.grid()
    plt.legend(['2018년-2019년 증감율', '2019년-2020년 증감률'], fontsize=15)
    
plt.subplot(5,1,5)
x = np.array(range(len(pct[4].columns)))
plt.bar(x-0.15, pct[4].loc[2019], width=w, label='2019')
plt.bar(x+0.15, pct[4].loc[2020], width=w, label='2018')
plt.xticks(range(11), pct[4].columns, rotation=25, fontsize=13)
plt.yticks(fontsize=20)
plt.ylabel('전년 대비 증감률 (%)', fontsize=20)
plt.legend(['2018년-2019년 증감율', '2019년-2020년 증감률'], fontsize=15)
plt.grid()
plt.show()
```
